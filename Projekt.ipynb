### 3.2.1 Pretvroba i objedinjenje podataka

Podatci iz datoteke `API_EN.ATM.NOXE.KT.CE_DS2_en_xml_v2_6302300` su u XML formatu. XML je jezik za označavanje koji se koristi za pohranu i prijenos podataka. XML datoteke se mogu pretvoriti u CSV format uz pomoć Python biblioteke `xml.etree.ElementTree`.

Nakon pretvorbe XML datoteke u CSV format, novu CSV datoteku ćemo spojiti sa datotekom `annual-co2-emissions-per-country.csv` uz pomoć naredbe `pd.merge()`. Ovom naredbom ćemo spojiti podatke o emisiji dušikovog oksida i podatke o ugljikovog dioksida u jednu datoteku.

````python
import xml.etree.ElementTree as ET

# Čitaj XML datoteku
tree = ET.parse('data/API_EN.ATM.NOXE.KT.CE_DS2_en_xml_v2_6302300.xml')
root = tree.getroot()

# EEstrahiraj podatke iz XML datoteke [9, StackOverflow]
data = []
for country in root.findall('.//record'):
    country_data = {}
    for child in country:
        country_data[child.attrib['name']] = child.text
    data.append(country_data)

# Stvori DataFrame iz XML podataka
df_xml = pd.DataFrame(data)

````

Kako bi se XML datoteka mogal pretvoriti u CSV format, potrebno ju je prvo pretvoriti u format koji je pandas-u razumljiv. To ćemo napraviti uz pomoć gore navedene naredbe `xml.etree.ElementTree`i For petlje. Nakon toga, podatci će biti spremljeni u DataFrame za daljnju obradu.

Dataframe je dalje moguće spremiti u CSV datoteku uz pomoć naredbe `to_csv()`.

````python
#Spremi podatke u CSV datoteku
df_xml.to_csv('data/API_EN.ATM.NOXE.KT.CE_DS2_en_xml_v2_6302300.csv')

# Uvezi CSV podatke
df_no_csv = pd.read_csv('data/API_EN.ATM.NOXE.KT.CE_DS2_en_xml_v2_6302300.csv')

# Prikaz podataka
print(df_no_csv)
````

### 5.2.2 Pohrana u bazu podataka

**Baza podataka je organizirana kolekcija podataka.**
Kako bi CSV ili JSON datoteku mogli spremiti u bazu podataka, potrebno je prvo kreirati bazu podataka. To ćemo napraviti uz pomoć Python biblioteke **sqlite3**.
>SQLite je softver koji implementira samostalnu, samodostatnu, nultu administraciju, transakcijsku SQL bazu podataka. [11]

Uz pomoć Pandas biblioteke, CSV datoteku ćemo pretvoriti u Dataframe. Nakon toga, Dataframe ćemo spremiti u bazu podataka uz pomoć naredbe `to_sql()`. Na vrlo elegantan način smo pretvorili CSV datoteku u lokalnu bazu podataka.

````python
?sqlite3
````

## 7.1 Korelacija i regresija

Korelacija je statistička mjera koja opisuje vezu između dvije varijable. Korelacija može biti pozitivna ili negativna. Pozitivna korelacija znači da kada jedna varijabla raste, druga varijabla također raste. Negativna korelacija znači da kada jedna varijabla raste, druga varijabla pada. Korelacija može biti i neutralna, što znači da ne postoji nikakva veza između varijabli.[8]

 Korelaciju ćemo koristiti za analizu veze između emisije stakleničkih plinova i porasta temperature na Zemlji. Da bi to prikazali koristit ćemo **scatter plot**.
 
  Scatter plot je graf koji prikazuje korelaciju između dvije varijable. Scatter plot se sastoji od točaka koje predstavljaju vrijednosti varijabli. Ako su točke raspoređene u obliku linije, to znači da postoji korelacija između varijabli. Ako su točke raspoređene nasumično, to znači da ne postoji korelacija između varijabli.

````python
df = pd.read_csv('data/merged_data4.csv')

#Filtriraj podatke za godine od 1990. do 2020.
df_filtered = df[(df['Godina'] >= 1990) & (df['Godina'] <= 2020)]

# Grupiraj podatke po godini i izračunaj ukupne emisije
df_grouped = df_filtered.groupby('Godina').sum()

# Nacrtaj graf raspršenja
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_grouped, x='Godišnje emisije plinova u CO₂ ekvivalentima', y='Promjena globalne srednje površinske temperature')
plt.xlabel('Ukupni staklenički plinovi (CO₂ ekvivalent)')
plt.ylabel('Promjena temperature (°C)')
plt.title('Korelacija između emisija plinova i promjene temperature (1990-2020) ')
plt.show()
````

Na prvi pogled, čini se da postoji pozitivna korelacija između emisije stakleničkih plinova i porasta temperature na Zemlji. To možemo provjeriti korsiteći **regresiju**. Jednostavno rečeno, regresija je linija koja najbolje odgovara podatcima. Regresija se koristi za predviđanje budućih vrijednosti.

````python
# Filtriraj podatke za godine od 1990. do 2020.
df_filtered = df[(df['Godina'] >= 1990) & (df['Godina'] <= 2020)]

# Grupiraj podatke po godini i izračunaj ukupne emisije
df_grouped = df_filtered.groupby('Godina').sum()

# Nacrtaj graf raspršenja s regresijskom linijom
plt.figure(figsize=(10, 6))
sns.regplot(data=df_grouped, x='Godišnje emisije plinova u CO₂ ekvivalentima', y='Promjena globalne srednje površinske temperature')
plt.xlabel('Ukupni staklenički plinovi (CO₂ ekvivalent)')
plt.ylabel('Promjena temperature (°C)')
plt.title('Korelacija između emisija plinova i promjene temperature (1990-2020) ')
plt.show()
````

Ovo smo mogli očekivati jer je poznato da emisija stakleničkih plinova uzrokuje globalno zatopljenje. Međutim, pogledajmo koleraciju između vrijednosti emisija ugljikovog dioksida i emisija dušikovog oksida.

````python
# Filtriraj podatke za godine od 1990. do 2020.
df_filtered = df[(df['Godina'] >= 1990) & (df['Godina'] <= 2020)]

# Grupiraj podatke po godini i izračunaj ukupne emisije
df_grouped = df_filtered.groupby('Godina').sum()

# Nacrtaj graf raspršenja s regresijskom linijom
plt.figure(figsize=(10, 6))
sns.regplot(data=df_grouped, x='Godišnje emisije CO₂', y='NOx emisije')
plt.xlabel('CO₂ emisije (t)')
plt.ylabel('NOx emisije (t)')
plt.title('Korelacija između emisija CO₂ i NOx (1990-2020) ')
plt.show()
````

````python
# Učitaj podatke
df = pd.read_csv('data/merged_data4.csv')

# Filtriraj podatke za sve države od 1990.
df_all_countries = df[df['Godina'] >= 1900]

# Grupiraj podatke po godini i državi i izračunaj ukupne emisije
df_grouped = df_all_countries.groupby(['Godina', 'Država']).sum().reset_index()

# Sumiraj vrijednosti za sve države
df_global = df_grouped.groupby('Godina').sum().reset_index()

# Pripremi podatke za linearnu regresiju
X = df_global['Godina'].values.reshape(-1, 1)
y = df_global['Godišnje emisije CO₂'].values

# Kreiraj model
model = LinearRegression()

# Treniraj model
model.fit(X, y)

# Napravi predikciju za 2050
future_years = range(1990, 2050)
future_X = pd.DataFrame({'Year': future_years})
future_y = model.predict(future_X)

# Nacrtaj graf predviđanja emisija CO2 za sve države do 2050
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Povijesni podaci')
plt.plot(future_X, future_y, color='red', label='Predviđanje')
plt.xlabel('Godina')
plt.ylabel('CO2 Emisije (t)')
plt.title('Predviđanje emisija CO2 za sve države do 2050.')
plt.legend()

# Prikazi graf
plt.tight_layout()
plt.show()
````

````python
# Kalkuliraj medijanu promjene temperature za svaku godinu
df_median = df_grouped.groupby('Godina')['Promjena globalne srednje površinske temperature'].median().reset_index()

# Pripremi podatke za linearnu regresiju
X_temp = df_median['Godina'].values.reshape(-1, 1)
y_temp = df_median['Promjena globalne srednje površinske temperature'].values

# Stvori model
model_temp = LinearRegression()

# Treiraj model
model_temp.fit(X_temp, y_temp)

# Napravi predikciju za 2050
future_years_temp = range(1990, 2050)
future_X_temp = pd.DataFrame({'Year': future_years_temp})
future_y_temp = model_temp.predict(future_X_temp)

# Nacrtaj graf predviđanja promjene temperature za sve države do 2050
plt.figure(figsize=(10, 6))
plt.scatter(X_temp, y_temp, color='blue', label='Historical data')
plt.plot(future_X_temp, future_y_temp, color='red', label='Prediction')
plt.xlabel('Year')
plt.ylabel('Temperature Change (°C)')
plt.title('Prediction of Global Temperature Change until 2050')
plt.legend()
plt.show()
````
